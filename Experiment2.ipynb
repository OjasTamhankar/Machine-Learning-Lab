##Univariate Analysis

simple descripive statistics and histograms for key variables like age, satisfactory_rating and monthly_income

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['age'], bins=10, kde=True)
plt.xlabel('Age')
plt.show()

sns.histplot(df['satisfaction_rating'], bins=10, kde=True)
plt.xlabel('Satisfaction Rating')
plt.show()

sns.histplot(df['monthly_income'], bins=10, kde=True)
plt.xlabel('Monthly Income')
plt.show()
```

##Bivariate Analysis

Lets check relationship between satisfactory level and monthly income

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='monthly_income', y='satisfaction_rating', data=df)
plt.title("Monthly Income vs Satisfaction Rating")
plt.xlabel("Monthly Income")
plt.ylabel("Satisfaction Level")
plt.show()
```

Group Analysis

analysing average satisfaction ratings by department and job role to see if certain roles or departments are most satisfied

```python
#Mean satisfaction by department
dept_satisfaction=df.groupby('department')['satisfaction_rating'].mean()
print("Average Satisfaction Rating by Dept\n", dept_satisfaction,"\n")
# Mean satisfaction by job role
role_satisfaction = df.groupby('job_role')['satisfaction_rating'].mean()
print("Average Satisfaction Rating by Job Role:\n", role_satisfaction)
```

```python
plt.pie(dept_satisfaction, labels=dept_satisfaction.index, autopct='%1.1f%%')
plt.title('Average Satisfaction Rating by Department')
plt.show()
```

```python
plt.bar(role_satisfaction.index, role_satisfaction.values)
plt.xlabel('Job Role')
plt.ylabel('Average Satisfaction Rating')
plt.title('Average Satisfaction Rating by Job Role')
plt.show()
```
